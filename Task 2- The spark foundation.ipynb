### Aim: From the given ‘Iris’ dataset, predict the optimum number of clusters and represent it visually.

### Dataset : https://bit.ly/3kXTdox

 **Using python to perform this task**

```python
# importing necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import warnings
warnings.filterwarnings("ignore")
```

```python
# load the Iris dataset
iris=load_iris()
x=iris.data[:,:2]   # consider only the first two parameters for clustering.
```

The code below performs k-means clustering for varying number of clusters (from 1 to 10) and compute the within-cluster sum of squares(WCSS) for each number of clusters. the result are plotted using Elbow method which helps to determine the optimal number of clusters.

###  Elbow method to find optimal number of clusters

```python
# Elbow method to find optimal number of clusters.
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
# Plotting the elbow curve
plt.plot(range(1,11),wcss)
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

As we can see from the plot, the "Elbow Point" located at k=3, which suggests that the optimal number of clusters for the iris dataset is 3.

Now we can perform K-means clustering with k=3, and visualize the resulting clusters using the following python code.

```python
# Kmeans clustering with k=3
kmeans=KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y=kmeans.fit_predict(x)
# Visualising the clusters
plt.scatter(x[pred_y==0,0], x[pred_y==0,1], s=100, c='red', label='Cluster1')
plt.scatter(x[pred_y==1,0], x[pred_y==1,1], s=100, c='blue', label='Cluster2')
plt.scatter(x[pred_y==2,0], x[pred_y==2,1], s=100, c='green', label='Cluster3')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s=300, c='yellow', label='Centroids')
plt.title('K-Means clustering')
plt.xlabel('Sepal Length(cm)')
plt.ylabel('Sepal Width(cm)')
plt.legend()
plt.show()
```

**The code above performs K-means clustering with k=3 and visualize the resulting clusters. each cluster is represented by a different colors, and centroids are shown in yellow.** 

As we can see from the plot, the data points have been divided into the 3 clusters.
